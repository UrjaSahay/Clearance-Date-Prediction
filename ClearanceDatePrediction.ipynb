```python
def handling_correlation(x_train_with_target,threshold=0.8):
    # creating set to hold the correlated features
    corr_features = set()
    # create the correlation matrix (default to pearson)
    corr_matrix = x_train_with_target.corr()
    #code for selecting all correleated features
    for i in range(len(corr_matrix .columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) >threshold:
                colname = corr_matrix.columns[i]
                corr_features.add(colname)
    return list(corr_features)
```

```python
def tree_based_feature_importance(x_train,y_train):
    from sklearn.ensemble import RandomForestRegressor
    # create the random forest model
    model = RandomForestRegressor(n_estimators=120)

    # fit the model to start training.
    model.fit(x_train, y_train)

    # get the importance of the resulting features.
    importances = model.feature_importances_

    # create a data frame for visualization.
    final_df = pd.DataFrame({"Features": x_train.columns, "Importances":importances})
    final_df.set_index('Importances')

    # sort in descending order 
    final_df = final_df.sort_values('Importances',ascending=False)
    
    #visualising feature importance
    pd.Series(model.feature_importances_, index=x_train.columns).nlargest(10).plot(kind='barh')  
    return final_df
```

```python
# the buckets that i will make for the values of the predicted target column 

#    <0 days
#    0 -15 days
#    16 -30 days
#    31 - 45 days
#    45 - 60 days
#    > 60 days
```

```python
# i have written a bucket function which will be applied to the columns to populate the column with their respective 
# aging buckets
```

```python
def bucket(a):
    if(a<0):
        return("< 0 ")
    elif((a >= 0) and (a<=15)):
        return("0 - 15")
    elif((a >= 16) and (a<=30)):
        return("16 - 30")
    elif((a >= 31) and (a <= 45)):
        return("31 - 45")
    elif((a >= 45) and (a <= 60)):
        return("45 - 60")
    else:
        return("> 60")
    
```
